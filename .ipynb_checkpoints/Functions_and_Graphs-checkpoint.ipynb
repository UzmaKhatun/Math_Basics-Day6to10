# üîπ Functions
## üî∏Definition
### A function is a rule that assigns each input value (from a set called the domain) to exactly one output value (from a set called the range).  In simpler terms, it's a relationship between inputs and outputs, where each input has only one output.

## Input: Also known as the independent variable or argument.
## Output: Also known as the dependent variable or the value of the function.

## Examples
## üìåLinear Function:
### Equation: ( f(x) = mx + c )
### Description: A straight-line relationship between ( x ) and ( f(x) ).  ( m ) is the slope, and ( c ) is the y-intercept.
### Example: ( f(x) = 2x + 3 )

## üìåQuadratic Function:
### Equation: ( f(x) = ax^2 + bx + c )
### Description: A parabolic relationship between ( x ) and ( f(x) ).
### Example: ( f(x) = x^2 - 4x + 4 )

## üìåPolynomial Function:
### Equation: ( f(x) = a_n x^n + a_{n-1} x^{n-1} + ... + a_1 x + a_0 )
### Description: A function that involves only non-negative integer powers of x.
### Example: ( f(x) = x^3 - 2x^2 + x - 7 )

## üìåExponential Function:
### Equation: ( f(x) = a^x ) (where ( a > 0 ) and ( a \neq 1 ))
### Description: The output increases rapidly as the input ( x ) increases.
### Example: ( f(x) = 2^x )

## üìåLogarithmic Function:
### Equation: ( f(x) = \log_a(x) ) (where ( a > 0 ) and ( a \neq 1 ))
### Description: The inverse of an exponential function.
### Example: ( f(x) = \log_{10}(x) )

## üìåTrigonometric Functions:
### Equations: ( f(x) = \sin(x) ), ( f(x) = \cos(x) ), ( f(x) = \tan(x) )
### Description: Functions that relate angles of a right triangle to ratios of its sides.  They are periodic functions.
### Examples: ( f(x) = \sin(x) )

## üìåActivation Functions:
### Equation: Various (see below)
### Description: In the context of neural networks, activation functions introduce non-linearity, allowing the network to learn complex patterns.
### Examples:
### Sigmoid: ( \sigma(x) = \frac{1}{1 + e^{-x}} )
### ReLU (Rectified Linear Unit): ( f(x) = \max(0, x) )
### Tanh (Hyperbolic Tangent): ( \tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}} )

## üîçWhy Functions Matter in Machine Learning
### - Modeling Relationships: Machine learning models learn functions that map input features to output predictions.
### - Algorithms: Many machine learning algorithms are based on optimizing functions (e.g., minimizing a cost function).
### - Feature Engineering: Functions are used to create new features from existing ones.
### - Non-linearity: Non-linear functions (like ReLU, sigmoid) are crucial for neural networks to learn complex patterns.

# üîπGraphs
## üî∏Definition
### In mathematics, a graph is a diagram that shows a relationship between two or more variables.  It typically consists of:
#### Axes: Usually, a horizontal x-axis and a vertical y-axis.
#### Points: Locations on the graph representing data points or values.
#### Lines or Curves: Used to connect the points and visualize the relationship between the variables.

## Types of Graphs
#### üìåLine Graph:
#### Used to show trends over time or the relationship between two continuous variables.
#### Example: A graph showing the change in stock price over a year.

### üìåBar Chart:
#### Used to compare categorical data.
#### Example: A graph comparing the sales of different products.

### üìåScatter Plot:
#### Used to show the relationship between two quantitative variables.  The pattern of the points reveals the type of relationship (e.g., positive, negative, or no correlation).
#### Example: A graph plotting height vs. weight for a group of people.

### üìåHistogram:
#### Used to show the distribution of a single quantitative variable.
#### Example: A graph showing the distribution of ages in a population.

## üî∏Graphs of Functions
### The graph of a function is a visual representation of the function's behavior.  If ( f ) is a function, its graph is the set of all points ( (x, f(x)) ) in the coordinate plane.
### For the linear function ( f(x) = mx + c ), the graph is a straight line.
### For the quadratic function ( f(x) = ax^2 + bx + c ), the graph is a parabola.

## üîçWhy Graphs Matter in Machine Learning
### Data Visualization: Graphs help visualize data, making it easier to understand patterns, trends, and outliers.
### Model Evaluation: Graphs are used to visualize the performance of machine learning models (e.g., plotting accuracy vs. training iterations, or ROC curves).
### Understanding Relationships: Graphs can reveal the relationships between input features and the target variable.
### Debugging: Graphs can help in diagnosing problems with a model or algorithm.
### Representing Complex Relationships: Beyond simple x-y graphs, more complex graph structures are used to represent relationships between data points, for example social networks, knowledge graphs, etc.

# Let's draw Graphs

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1. Linear Function
## Equation: f(x) = mx + c

```python
x = np.linspace(-10, 10, 100)
y = 2*x + 3

plt.plot(x, y, label='f(x) = 2x + 3')
plt.title("Linear Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

## 2. Quadratic Function
## Equation: f(x) = ax¬≤ + bx + c

```python
y = x**2 - 4*x + 3

plt.plot(x, y, label='f(x) = x¬≤ - 4x + 3')
plt.title("Quadratic Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

## 3. Exponential Function
## Equation: f(x) = a^x

```python
y = 2**x

plt.plot(x, y, label='f(x) = 2^x')
plt.title("Exponential Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

## Graphs help visualize model performance.  For example, a graph of "Accuracy vs. Training Iterations" shows how the model's accuracy improves over time.  If the graph plateaus, it suggests the model has stopped learning.  An ROC curve (Receiver Operating Characteristic curve) is used to visualize the performance of a classification model, showing the trade-off between true positive rate and false positive rate.
